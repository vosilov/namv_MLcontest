```python
def makeDicts(row):
    if len(row[0]) != 0:
        x = {row.index[1] : row[1],
             row.index[0] : row[0][:len(row[0])]}
    else:
        x = np.NaN
    return x
```

```python
keywordsDF.iloc[:,1] = keywordsDF.iloc[:,[1, 0]].apply(makeDicts, axis=1)
```

```python
keywordsDF.dropna(inplace=True)
display(keywordsDF)
```
